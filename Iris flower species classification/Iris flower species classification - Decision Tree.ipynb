## Import Libraries

### Segregating Target & Predictors

```python
# independent variables
X = data.drop(['target_species', 'species_names'], axis = 1)

# target/dependent variable
Y = data['target_species']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

# spliting the data into train & test set (75:25 ratio)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)

# checking the shape of train & test set

print("Shape of train set :")
print(X_train.shape, Y_train.shape)
print("-"*50)
print("Shape of test set :")
print(X_test.shape, Y_test.shape)
```

### Model Building - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Creating an instance of decision tree model
model = DecisionTreeClassifier(random_state = 0)

# fit on training data
model.fit(X_train,Y_train)
```

### Model Prediction

```python
# making prediction on test set
y_pred = model.predict(X_test)
y_pred
```

### Model Evaluation

**Accuracy Score**

```python
print("Accuracy of the Model: {0}%".format(round(accuracy_score(Y_test, y_pred),2)*100))
```

**confusion Matrix**

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cf_matrix = confusion_matrix(Y_test, y_pred)
cf_matrix
```

```python
# visualising confusion matrix using heatmap
plt.figure(figsize = (6,6))
sns.heatmap(cf_matrix, annot = True, fmt = 'g')
plt.title("Confusion Matrix", fontsize = 15, weight = 'bold')
plt.xlabel("Predicted Values", fontsize = 15)
plt.ylabel("`Actual Values", fontsize = 15)
plt.show()
```

**Classification Report**

```python
# Classification report 
print(classification_report(Y_test, y_pred))
```

### Model Improvement

#### Finding Best Max_depth Value

```python
accuracies = []
for i in range(1, 11):
    model = DecisionTreeClassifier(max_depth = i, random_state = 0)
    model.fit(X_train, Y_train)
    preds = model.predict(X_test)
    score = accuracy_score(Y_test, preds)
    accuracies.append(score)

plt.figure(figsize=(12, 6))
plt.plot(range(1, 11),
         accuracies,
         color='red',
         linestyle='dashed',
         marker='o',
         markerfacecolor='blue',
         markersize =10)

plt.title('Finding best Max_Depth Value')
plt.xlabel('max_depth values')
plt.ylabel('Accuracy scores')   
plt.show()
```

```python
#  decision tree model using max_depth value =3
model = DecisionTreeClassifier(max_depth = 3, random_state = 0)

# fit on training data
model.fit(X_train,Y_train)

# making prediction on test set
y_pred = model.predict(X_test)

# calculating accuracy of the model
print("Accuracy of the Model: {0}%".format(round(accuracy_score(Y_test, y_pred),2)*100))
```

**Changing the ratio of Train & test data (80:20)**

```python
# independent variables
X = data.drop(['target_species', 'species_names'], axis = 1)

# target/dependent variable
Y = data['target_species']
```

```python
from sklearn.model_selection import train_test_split

# spliting the data into train & test set (80:20 ratio)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)

# checking the shape of train & test set

print("Shape of train set :")
print(X_train.shape, Y_train.shape)
print("-"*50)
print("Shape of test set :")
print(X_test.shape, Y_test.shape)
```

```python
#  decision tree model 
best_model = DecisionTreeClassifier(random_state = 0)

# fit on training data
best_model.fit(X_train,Y_train)

# making prediction on test set
y_pred = best_model.predict(X_test)

# calculating accuracy of the model
print("Accuracy of the Model is: {0}%".format(round(accuracy_score(Y_test, y_pred),2)*100))
```

#### Observations:
+ After changing the ratio of train & test set to 80:20 ,accuracy of the model increased to 100% from 97%.
